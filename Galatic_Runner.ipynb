```python
import pygame
from sys import exit
from random import randint
import os

def display_score():
    # Calcula el tiempo actual desde el inicio del juego y muestra la puntuación
    current_time = int(pygame.time.get_ticks()/1000) - start_time
    score_surf = text_font.render(f'Score:{current_time}', False, ((255,255,240)))
    score_rect = score_surf.get_rect(center=(400,150))
    screen.blit(score_surf, score_rect)
    return current_time

def obstacle_movement(obstacle_list):
    # Mueve los obstáculos hacia la izquierda y elimina los que salen de la pantalla
    if obstacle_list:
        for obstacle_rect in obstacle_list:
            obstacle_rect.x -= 5
            
            if obstacle_rect.bottom == 590:
                screen.blit(monster_surf, obstacle_rect)
            else:
                screen.blit(fly_surf, obstacle_rect)
            
        obstacle_list =[obstacle for obstacle in obstacle_list if obstacle.x > -100]
        
        return obstacle_list
    else:
        return []
    
def collisions(player, obstacles):
    # Verifica si el jugador colisiona con algún obstáculo
    if obstacles:
        for obstacle_rect in obstacles:
            if player.colliderect(obstacle_rect): return False
    return True
                
                
def player_animation():
    # Anima al jugador entre correr y saltar
    global player_surf, player_index
    
    if player_rect.bottom < 300:
        player_surf = player_jump
    else:
        player_index += 0.1
        if player_index >= len(player_walk): player_index = 0
        player_surf = player_walk[int(player_index)]

# Inicialización de Pygame y configuración de la ventana del juego
pygame.init() # Inicializa todos los módulos de pygame
screen = pygame.display.set_mode((800,600)) # Crea una ventana de 800x600
pygame.display.set_caption('Galactic Runner') # Establece el título de la ventana
clock = pygame.time.Clock() # Crea un objeto para controlar el tiempo
text_font = pygame.font.Font('C:/Users/USUARIO/Kodland/PlayfulTime-BLBB8.ttf', 65) # Carga de la fuente de texto

# Variables de estado inicial del juego
game_active = False
start_time= 0
score = 0

# Carga y configuración de imágenes para el fondo y el suelo
sky_surface = pygame.image.load('C:/Users/USUARIO/Kodland/sky_surface.jpg').convert()
ground_surface = pygame.image.load('C:/Users/USUARIO/Kodland/groound.png').convert_alpha()

# Configuración inicial de la pantalla y del texto del juego
score_surf = text_font.render('Galactic Runner', False, (255,255,240))
score_rect = score_surf.get_rect(center = (400, 150))

# Carga y configuración de los obstáculos (monstruos y asteroides)
monster_frame_1 = pygame.image.load('C:/Users/USUARIO/Kodland/moonsteer.png').convert_alpha()
monster_frame_2 = pygame.image.load('C:/Users/USUARIO/Kodland/moonsteer.png').convert_alpha()
monster_frames = [monster_frame_1, monster_frame_2]
monster_frame_index = 0
monster_surf = monster_frames[monster_frame_index]

fly_frame1 = pygame.image.load('C:/Users/USUARIO/Kodland/fly.png').convert_alpha()
fly_frame2 = pygame.image.load('C:/Users/USUARIO/Kodland/fly.png').convert_alpha()
fly_frames = [fly_frame1, fly_frame2]
fly_frame_index = 0
fly_surf = fly_frames[fly_frame_index]

obstacle_rect_list =[]

# Carga y configuración del jugador (animación de caminar y saltar)
player_walk1 = pygame.image.load('C:/Users/USUARIO/Kodland/walk1.png').convert_alpha()
player_walk2 = pygame.image.load('C:/Users/USUARIO/Kodland/walk2.png').convert_alpha()
player_walk = [player_walk1, player_walk2]
player_index = 0
player_jump = pygame.image.load('C:/Users/USUARIO/Kodland/astronautaa.png').convert_alpha()

player_surf = player_walk[player_index]
player_rect = player_surf.get_rect(midbottom =(50,580))
player_gravity = 0

# Pantalla de introducción
player_stand = pygame.image.load('C:/Users/USUARIO/Kodland/gameover.png').convert_alpha()
player_stand_rect = player_stand.get_rect(center=(400,300))

game_name = text_font.render('Galactic Runner', False, (255,255,240))
game_name_rect = game_name.get_rect(center=(400,100))

game_message = text_font.render('Press space to run', False, (255,255,240))
game_message_rect = game_message.get_rect(center=(400,530))

#Temporizadores
obstacle_timer = pygame.USEREVENT + 1
pygame.time.set_timer(obstacle_timer, 2000)

monster_animation_timer = pygame.USEREVENT + 2
pygame.time.set_timer(monster_animation_timer, 500)

fly_animation_timer = pygame.USEREVENT + 3
pygame.time.set_timer(fly_animation_timer, 200)

# Bucle principal del juego
while True:
    for event in pygame.event.get():
        # Cierra el juego de manera limpia si el usuario desea salir
        if event.type == pygame.QUIT:
            pygame.quit()
            exit()
            
        if game_active: 
            if event.type == pygame.MOUSEMOTION:
                 if player_rect.collidepoint(event.pos) and player_rect.bottom >= 575:
                        player_gravity= -20

            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE and player_rect.bottom >= 575:
                    player_gravity = -20
        else:
            if event.type == pygame.KEYDOWN and event.key == pygame.K_SPACE:
                game_active = True
                
                start_time = int(pygame.time.get_ticks()/1000)
                
        # Gestiona eventos específicos del juego cuando está activo        
        if game_active:
            if event.type == obstacle_timer:
                if randint(0,2):
                    obstacle_rect_list.append(fly_surf.get_rect(bottomright = (randint(900,1100),590)))
                else:
                    obstacle_rect_list.append(monster_surf.get_rect(bottomright = (randint(900,1100),400)))
            
            # Cambia el cuadro de animación del monstruo basado en un temporizador
            if event.type == monster_animation_timer:
                if monster_frame_index == 0: monster_frame_index = 1
                else: monster_frame_index = 0
                monster_surf = monster_frames[monster_frame_index]

                
            # Cambia el cuadro de animación del asteroide basado en un temporizador
            if event.type == fly_animation_timer:
                if fly_frame_index == 0: fly_frame_index = 1
                else: fly_frame_index = 0
                fly_surf = fly_frames[fly_frame_index]
                
    # Si el juego está activo, actualiza y dibuja todos los elementos del juego
    if game_active:
        screen.blit(sky_surface, (0,0))
        screen.blit(ground_surface, (0,550))
        score = display_score()

        #Actualiza la gravedad del jugador, lo que afecta su salto
        player_gravity += 1
        player_rect.y += player_gravity
        if player_rect.bottom >= 575: player_rect.bottom = 575
        player_animation()
        screen.blit(player_surf, player_rect)
        
        # Mueve los obstáculos y los dibuja
        obstacle_rect_list = obstacle_movement(obstacle_rect_list)

        # Comprueba las colisiones, lo que puede desactivar el juego
        game_active = collisions(player_rect,obstacle_rect_list)
        
    # Muestra la pantalla de inicio o de game over        
    else:
        screen.fill((0,0,0))
        screen.blit(player_stand , player_stand_rect)
        obstacle_rect_list.clear()
        player_rect.midbottom = (50,580)
        player_gravity = 0
        
        # Muestra el mensaje de puntuación o de inicio
        score_message = text_font.render(f'Your score: {score}', False, (255,255,240))
        score_message_rect = score_message.get_rect(center=(400,530))
        screen.blit(game_name, game_name_rect)
        
        if score == 0: screen.blit(game_message, game_message_rect)
        else: screen.blit(score_message, score_message_rect)
    
    # Actualiza la pantalla con todo lo dibujado
    pygame.display.update()
    # Mantiene el juego corriendo a 60 cuadros por segundo
    clock.tick(60)
```
